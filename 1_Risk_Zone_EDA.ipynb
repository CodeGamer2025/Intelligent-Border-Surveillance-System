```python
import pandas as pd

# 1. Define the exact path to your ACLED dataset
file_path = "Datasets/ACLED (Armed Conflict Location & Event Data Project)/ACLED Data_2026-03-10.csv"

# 2. Load the CSV file into a Pandas DataFrame
df_acled = pd.read_csv(file_path)

# 3. Check how big the dataset is (Rows, Columns)
print(f"Total Rows and Columns: {df_acled.shape}")

# 4. Display the first 5 rows
df_acled.head()
```

```python
# 1. Keep only the columns needed for Risk Zone mapping
columns_to_keep = ['event_date', 'year', 'event_type', 'country', 'latitude', 'longitude', 'fatalities']
df_clean = df_acled[columns_to_keep].copy()

# 2. Drop any rows that are missing GPS coordinates (we can't map them without it!)
df_clean = df_clean.dropna(subset=['latitude', 'longitude'])

print(f"Cleaned Data ready for mapping: {df_clean.shape[0]} rows.")
df_clean.head()
```

```python
import folium
from folium.plugins import HeatMap

# 1. Create a base map centered around the Indian subcontinent
border_map = folium.Map(location=[22.0, 78.0], zoom_start=5, tiles='cartodb dark_matter')

# 2. Extract latitude and longitude into a list for the heatmap
heat_data = [[row['latitude'], row['longitude']] for index, row in df_clean.iterrows()]

# 3. Add the heatmap layer to the base map
HeatMap(heat_data, radius=8, blur=12, max_zoom=1).add_to(border_map)

# 4. Display the map right here in the notebook
border_map
```

```python
# 1. Filter the dataset to ONLY include events inside India
df_india_only = df_clean[df_clean['country'] == 'India'].copy()

print(f"Rows remaining after filtering for India: {df_india_only.shape[0]}")

# 2. Create a new base map
india_border_map = folium.Map(location=[22.0, 78.0], zoom_start=5, tiles='cartodb dark_matter')

# 3. Extract the new filtered coordinates
heat_data_india = [[row['latitude'], row['longitude']] for index, row in df_india_only.iterrows()]

# 4. Add the heatmap layer
HeatMap(heat_data_india, radius=8, blur=12, max_zoom=1).add_to(india_border_map)

# 5. Display the cleaned-up map
india_border_map
```
